```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
dummy_time = np.arange(0.001, 1.3e4, 100) # time from beginning of universe to 13 Gyrs later. unit: Myr
```

```python
def sfhdelayed(t, tau):
    SFR = (t/(tau)**2) * np.exp(-t/tau)
    return SFR
```

```python
def sfh2exp(t, t1, t2, tau1, tau2, k=1):
    SFR = []
    for time in t:
        if time < t1 - t2:
            SFR = np.append(SFR, np.exp(-time/tau1))
        else:
            SFR = np.append(SFR, np.exp(-time/tau1) + k*np.exp(-time/tau2))
    return SFR
```

```python
mod2exp1 = sfh2exp(dummy_time, 10000, 1050, 7000, 100)
mod2exp2 = sfh2exp(dummy_time, 10000, 0, 7000, 90)
moddelayed1 = sfhdelayed(dummy_time, 1000)
moddelayed2 = sfhdelayed(dummy_time, 7000)
```

```python
plt.plot(dummy_time, mod2exp1, label='sfh2exp, tau=1000')
plt.plot(dummy_time, mod2exp2 * (max(mod2exp1) / max(mod2exp2)), label='sfh2exp, tau=7000')
plt.plot(dummy_time, moddelayed1 * (max(mod2exp1) / max(moddelayed1)), label='sfhdelayed, tau=1000')
plt.plot(dummy_time, moddelayed2 * (max(mod2exp1) / max(moddelayed2)), label='sfhdelayed, tau=7000')
plt.legend()
plt.show()
```

```python
def sfh2exp(t, t1, t2, tau1, tau2):
    SFR = []
    for time in t:
        if time < t1 - t2:
            SFR = np.append(SFR, np.exp(-time/tau1))
        else:
            SFR = np.append(SFR, np.exp(-time/tau1) + np.exp(-time/tau2))
    return SFR
```

```python
sampsize = 5

t1samp = np.random.choice(dummy_time[1:50], sampsize)
t2samp = np.random.choice(dummy_time[(dummy_time > 10) & (dummy_time < 300)], sampsize)
tau1samp = np.random.uniform(250,9000,sampsize)
tau2samp = np.random.uniform(1,50,sampsize)

expsamps = [sfh2exp(dummy_time, t1, t2, tau1, tau2, k=60) for (t1, t2, tau1, tau2) in zip(t1samp, t2samp, tau1samp, tau2samp)]
```

```python
for i in range(sampsize):
    notation = 't1=' + str(np.round(t1samp[i])) + ', t2=' + str(np.round(t2samp[i])) + \
            ', tau1=' + str(np.round(tau1samp[i])) + ', tau2=' + str(np.round(tau2samp[i]))
    plt.plot(dummy_time, expsamps[i], label=notation)
    plt.legend()
```
