```python
import numpy as np
import pandas as pd
import astropy.units as u

from astropy.io import fits
from astropy.wcs import WCS
from astropy.wcs.utils import pixel_to_skycoord
from astropy.utils.data import get_pkg_data_filename
from astropy.coordinates import SkyCoord, Angle

from regions import Regions, CircleSkyRegion

catalog_file = 'COSMOS-Web_supercatalog_v0.7.fits'

top_left = pixel_to_skycoord(765.85, 15404.9, wcs)
top_right = pixel_to_skycoord(6838.53, 17778.8, wcs)
bot_left = pixel_to_skycoord(5025.03, 4634.77, wcs)
bot_right = pixel_to_skycoord(11117.3, 7004.25, wcs)

hdu_cat = fits.open(catalog_file)
hdu_cat.info()

cat_data = hdu_cat[1].data
hdu_cat.close()

cat_coords = []

i = 0
while i < 100000:
    cat_ra = cat_data[i][1]*u.deg
    cat_dec = cat_data[i][2]*u.deg
    if Angle(bot_right.ra) < Angle(cat_ra) < Angle(top_left.ra) and Angle(bot_left.dec) < Angle(cat_dec) < Angle(top_right.dec):
        cat_coords.append(SkyCoord(cat_ra.value, cat_dec.value, unit="deg"))
    i += 1

regs = Regions([CircleSkyRegion(cat_coords[i], 1.5*u.arcsec) for i in range(len(cat_coords))])
regs.write('catalog_coords.reg', format='ds9')
```
